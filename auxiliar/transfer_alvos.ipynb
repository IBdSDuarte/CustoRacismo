## Casos por ano

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que 'alvos' seja o DataFrame
alvos["Ano"] = alvos["Data"].str[-2:]  # Extrai os anos das datas

# Contagem de casos por ano e tipo de discriminação
casos_por_ano = alvos.groupby(["Ano", "Tipo"]).size().unstack(fill_value=0)

# Plotando o gráfico
plt.figure(figsize=(24, 12))
for tipo in casos_por_ano.columns:
    plt.plot(casos_por_ano.index, casos_por_ano[tipo], marker='o', label=tipo)

plt.xlabel("Ano")
plt.ylabel("Número de Casos")
plt.title("Casos de Racismo, LGBTfobia e Xenofobia por Ano")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

## Marca com true

```python
transferencias.insert(4, 'Xeno', False)  # O "1" especifica a posição após a coluna 'Nome'
transferencias.insert(4, 'LGBT', False)  # O "1" especifica a posição após a coluna 'Nome'
transferencias.insert(4, 'Raci', False)  # O "1" especifica a posição após a coluna 'Nome'

transferencias['Alvo'] = transferencias['ID'].isin(alvos['UID_TMK'])
```

```python
# Iterar sobre os tipos na tabela alvos
for index, row in alvos.iterrows():
    # Obter o tipo de violência do alvo
    tipo_violencia = row['Tipo']
    
    # Encontrar o ID correspondente ao alvo na tabela transferencias
    id_vitima = row['UID_TMK']
    
    # Atribuir os valores booleanos nas colunas corretas de acordo com o tipo de violência
    if tipo_violencia == 'Racismo':
        transferencias.loc[transferencias['ID'] == id_vitima, 'Raci'] = True
    elif tipo_violencia == 'LGBTfobia':
        transferencias.loc[transferencias['ID'] == id_vitima, 'LGBT'] = True
    elif tipo_violencia == 'Racismo/Xenofobia':
        transferencias.loc[transferencias['ID'] == id_vitima, 'Raci'] = True
        transferencias.loc[transferencias['ID'] == id_vitima, 'Xeno'] = True
    elif tipo_violencia == 'Xenofobia':
        transferencias.loc[transferencias['ID'] == id_vitima, 'Xeno'] = True
```

# Contagem das transferencia-alvo

```python
# Contar quantos True existem na coluna 'Alvo'
qtd_alvo = transferencias['Alvo'].sum()
print('Alvo\t\t',qtd_alvo)

# Contar quantos False existem na coluna 'Alvo'
qtd_n_alvo = (transferencias['Alvo'] == False).sum()
print('Não alvo\t',qtd_n_alvo)

# Contar quantos casos de xenofobia foram registrados
qtd_xeno = transferencias['Xeno'].sum()
print('Xenofobia\t', qtd_xeno)

# Contar quantos casos de racismo foram registrados
qtd_raci = transferencias['Raci'].sum()
print('Racismo\t\t',qtd_raci)

# Contar quantos casos de homofobia foram registrados
qtd_lgbt = transferencias['LGBT'].sum()
print('LGBT\t\t',qtd_lgbt)
```

```python
subconjunto_xeno
```

```python
# Filtrando os casos de xenofobia na tabela 'alvos'
subconjunto_xeno = alvos[alvos['Tipo'] == 'Xenofobia']

# Inicializando uma lista para armazenar os jogadores
jogadores_xeno = []

# Para cada caso de xenofobia, verificamos se o jogador está na tabela de transferências
for _, linha in subconjunto_xeno.iterrows():
    jogador = transferencias[transferencias['ID'] == linha['UID_TMK']]['Nome'].values
    if jogador.size > 0:
        jogadores_xeno.append(jogador[0])

# Contando quantas vezes cada jogador foi vítima de xenofobia
contagem_xeno = pd.Series(jogadores_xeno).value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
contagem_xeno.plot(kind="bar", color="red")
plt.xlabel("Jogador")
plt.ylabel("Número de Casos")
plt.title("Casos de Xenofobia por Jogador")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

## 2016 - OK

```python
# Carregar o arquivo YAML
yaml_file = '2016.yaml'  # Substitua pelo caminho correto do seu arquivo

with open(yaml_file, 'r', encoding='utf-8') as file:
    data = yaml.safe_load(file)

# Criar um DataFrame a partir dos dados YAML
df_16 = pd.DataFrame(data)

# Exibir o DataFrame
#df_16.head(10)
#df_16['Tipo'].unique()
#df_16['Campeonato'].unique()

# Função para atribuir 'Internet' ou 'Físico'
def definir_ambiente(campeonato):
    if 'Internet' in str(campeonato):  # Convertendo o valor para string
        return 'Internet'
    else:
        return 'Físico'

# Garantir que não haja valores NaN e que todos os valores sejam strings
df_16['Campeonato'] = df_16['Campeonato'].fillna('')

# Criar a nova coluna 'Ambiente' baseado na coluna 'Campeonato'
df_16['Ambiente'] = df_16['Campeonato'].apply(definir_ambiente)

# Mover a coluna 'Ambiente' para ficar após a coluna 'Título'
colunas = df_16.columns.tolist()
colunas.insert(colunas.index('Título') + 1, colunas.pop(colunas.index('Ambiente')))
df_16 = df_16[colunas]

 #Cria as colunas 'Nome' e 'Clube' a partir da coluna 'Título'
df_16[['Vítima', 'Clube_Vit']] = df_16['Título'].str.split(r'[\–\-]', n=1, expand=True)

# Remove espaços em branco desnecessários
df_16['Vítima'] = df_16['Vítima'].str.strip()
df_16['Clube_Vit'] = df_16['Clube_Vit'].str.strip()

# Exibir o DataFrame com as novas colunas
df_16.head(2)
```

```python
df_16_atletas = df_16
df_16_atletas = df_16_atletas[~df_16_atletas['Vítima'].str.contains('Torcedor', case=False, na=False)]
df_16_atletas = df_16_atletas[~df_16_atletas['Vítima'].str.contains('Torcida', case=False, na=False)]
df_16_atletas = df_16_atletas[~df_16_atletas['Vítima'].str.contains('Políticos', case=False, na=False)]
df_16_atletas = df_16_atletas[~df_16_atletas['Vítima'].str.contains('Janaina', case=False, na=False)]
df_16_atletas = df_16_atletas[~df_16_atletas['Vítima'].str.contains('Marta', case=False, na=False)]
df_16_atletas = df_16_atletas[~df_16_atletas['Vítima'].str.contains('Juan Carlos', case=False, na=False)]
df_16_atletas = df_16_atletas[~df_16_atletas['Vítima'].str.contains('Equipe', case=False, na=False)]

df_16_atletas = df_16_atletas[~df_16_atletas['Clube_Vit'].str.contains('Árbitro', case=False, na=False)]
df_16_atletas = df_16_atletas[~df_16_atletas['Clube_Vit'].str.contains('Treinador', case=False, na=False)]
df_16_atletas = df_16_atletas[~df_16_atletas['Clube_Vit'].str.contains('Handbol', case=False, na=False)]


# Exibir o novo DataFrame
df_16_atletas
```

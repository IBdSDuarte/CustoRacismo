```python
# Carregar o arquivo YAML
yaml_file = '2015.yaml'  # Substitua pelo caminho correto do seu arquivo

with open(yaml_file, 'r', encoding='utf-8') as file:
    data = yaml.safe_load(file)

# Criar um DataFrame a partir dos dados YAML
df_15 = pd.DataFrame(data)

# Selecionar as linhas que contêm 'Atleta' no título
df_15_atletas = df_15[df_15['Título'].str.contains('Atleta', case=False, na=False)].copy()
# Remover 'Atleta' da coluna 'Título' usando .loc para evitar o warning
df_15_atletas.loc[:, 'Título'] = df_15_atletas['Título'].str.replace('Atleta', '', case=False).str.strip()
df_15_atletas.loc[:, 'Título'] = df_15_atletas['Título'].str.replace('Caso', '', case=False).str.strip()

# Divide a coluna 'Título' com base nas vírgulas
df_15_atletas[['Vítima', 'Clube_Vit']] = df_15_atletas['Título'].str.split(',', n=1, expand=True)

# Remove espaços em branco desnecessários
df_15_atletas['Vítima'] = df_15_atletas['Vítima'].str.strip()
df_15_atletas.loc[:, 'Clube_Vit'] = df_15_atletas['Clube_Vit'].str.replace(',', '', case=False).str.strip()
df_15_atletas['Clube_Vit'] = df_15_atletas['Clube_Vit'].str.strip()
```

```python
# Carregar o arquivo YAML
yaml_file = '2016.yaml'  # Substitua pelo caminho correto do seu arquivo

with open(yaml_file, 'r', encoding='utf-8') as file:
    data = yaml.safe_load(file)

# Criar um DataFrame a partir dos dados YAML
df_16 = pd.DataFrame(data)

# Função para atribuir 'Internet' ou 'Físico'
def definir_ambiente(campeonato):
    if 'Internet' in str(campeonato):  # Convertendo o valor para string
        return 'Internet'
    else:
        return 'Físico'

# Garantir que não haja valores NaN e que todos os valores sejam strings
df_16['Campeonato'] = df_16['Campeonato'].fillna('')

# Criar a nova coluna 'Ambiente' baseado na coluna 'Campeonato'
df_16['Ambiente'] = df_16['Campeonato'].apply(definir_ambiente)

# Mover a coluna 'Ambiente' para ficar após a coluna 'Título'
colunas = df_16.columns.tolist()
colunas.insert(colunas.index('Título') + 1, colunas.pop(colunas.index('Ambiente')))
df_16 = df_16[colunas]

 #Cria as colunas 'Nome' e 'Clube' a partir da coluna 'Título'
df_16[['Vítima', 'Clube_Vit']] = df_16['Título'].str.split(r'[\–\-]', n=1, expand=True)

# Remove espaços em branco desnecessários
df_16['Vítima'] = df_16['Vítima'].str.strip()
df_16['Clube_Vit'] = df_16['Clube_Vit'].str.strip()
```

```python
# Destacar colunas onde há NaN em qualquer posição
def highlight_missing(s):
    return ['background-color: yellow' if pd.isna(v) else '' for v in s]

styled_df = df_types.style.apply(highlight_missing, axis=1)
styled_df
```
